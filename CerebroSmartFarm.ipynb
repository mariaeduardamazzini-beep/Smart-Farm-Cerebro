Download do Dataset

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("abdulahad0296/indoor-plant-disease-detection-dataset")

print("Path to dataset files:", path)
```

Imports

```python
import tensorflow as tf
from tensorflow.keras import models, layers
```

ExtraГ§ГЈo de imagens do dataset

```python
#config. das imagens
img_height, img_width = 180, 180
batch_size = 32

# Treino do dataset
train_ds_plantascaseiras = tf.keras.utils.image_dataset_from_directory(
  path,
  validation_split=0.0, #0.0 - 100% dos dados para o treino
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
# ValidaГ§ГЈo
val_ds = tf.keras.utils.image_dataset_from_directory(
  path,
  validation_split=0.3, #30% dos dados para validaГ§ГЈo
  subset="validation",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

ConstruГ§ГЈo do Modelo Convolucional (CNN) - ClassificaГ§ГЈo e DetecГ§ГЈo de Imagens

```python
num_classes = len(train_ds_plantascaseiras.class_names)

model = models.Sequential([
  #Camada de redimensionamento
  layers.Rescaling(1./255, input_shape=(img_height, img_width, 3)),

  #Camada Convolucionais - caracterГ­stica das imagens
  layers.Conv2D(32, 3, activation='relu'),
  layers.MaxPooling2D(),

  layers.Conv2D(64, 3, activation='relu'),
  layers.MaxPooling2D(),

  #Camada Flatten - transforma pixels em vetor
  layers.Flatten(),

  #Camada Dense - decisГЈo final
  layers.Dense(128, activation='relu'),
  layers.Dense(num_classes, activation='softmax')
])

model.compile(
  optimizer='adam',
  loss=tf.keras.losses.SparseCategoricalCrossentropy(),
  metrics=['accuracy']
)
```

```python
# Treimento da IA e validaГ§ГЈo (100% correspondente ao treino e 30% a validaГ§ГЈo)
history = model.fit(
  train_ds_plantascaseiras,
  validation_data=val_ds,
  epochs=20
)
```
